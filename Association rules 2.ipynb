```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder

# Load the dataset
df = pd.read_excel(r"C:\Users\acer\Downloads\Association Rules\Association Rules\Online retail.xlsx", sheet_name="Sheet1", engine="openpyxl")

# Display basic information
print(df.info())

# Data Preprocessing
print("Columns in dataset:", df.columns)
df.columns = ["Items"]  # Rename the single column to "Items"

# Convert transactions into a list of lists
transactions = df["Items"].apply(lambda x: x.split(","))

# Convert transactions into a one-hot encoded DataFrame
te = TransactionEncoder()
basket_array = te.fit(transactions).transform(transactions)
basket = pd.DataFrame(basket_array, columns=te.columns_)

# Apply Apriori Algorithm with expanded range of support and confidence values
support_values = np.array([0.005, 0.01, 0.02, 0.05, 0.1])
confidence_values = np.array([0.1, 0.2, 0.4, 0.6, 0.8])

for sup in support_values:
    frequent_itemsets = apriori(basket, min_support=sup, use_colnames=True)

    print(f"\nFrequent Itemsets for Support = {sup}")
    print(frequent_itemsets.head())

    for conf in confidence_values:
        rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
        rules = rules[(rules['confidence'] >= conf)]
        
        print(f"\nAssociation Rules for Support = {sup} and Confidence = {conf}")
        print(rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']].head())

        # Visualization of Lift vs Confidence
        plt.figure(figsize=(8, 5))
        sns.scatterplot(x=rules['confidence'], y=rules['lift'], size=rules['support'], alpha=0.7)
        plt.xlabel('Confidence')
        plt.ylabel('Lift')
        plt.title(f'Lift vs Confidence (Support={sup}, Confidence={conf})')
        plt.show()

        # Visualization of Support vs Confidence
        plt.figure(figsize=(8, 5))
        sns.scatterplot(x=rules['support'], y=rules['confidence'], size=rules['lift'], alpha=0.7)
        plt.xlabel('Support')
        plt.ylabel('Confidence')
        plt.title(f'Support vs Confidence (Support={sup}, Confidence={conf})')
        plt.show()

        # Heatmap for correlation between support, confidence, and lift
        plt.figure(figsize=(8, 5))
        sns.heatmap(rules[['support', 'confidence', 'lift']].corr(), annot=True, cmap='coolwarm', fmt='.2f')
        plt.title(f'Correlation Heatmap (Support={sup}, Confidence={conf})')
        plt.show()
```
